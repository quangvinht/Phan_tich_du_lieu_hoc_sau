```python
data = {
        'sharesz':[5,9,6,8.5,3,13.6,4.6,6.7,3.0,7.7],
        'expectedz':[15,14,15,17,11,19,13,14,10,13]
}

dff = pd.DataFrame(data = data)
dff
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr
import statsmodels.formula.api as smf
```

```python
# V scatter plot , X trc y sau :
plt.plot(dff['sharesz'],dff['expectedz'],'o')
plt.xlabel('Shares Selling')
plt.ylabel('Expected Price')
plt.title('Shares Selling v Expected Price')
plt.show()
```

```python
# tnh h s tng quan :
print('H s tng quan l :',pearsonr(dff['sharesz'],dff['expectedz']))
```

```python
# Tm phng trnh hi quy : tm beta0 , beta1
#truyn Y trc : presz ,  X sau :postz
result = smf.ols('expectedz~sharesz',dff).fit()
result.summary()
```
